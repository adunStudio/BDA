```python
"""
Question 6

ëŒ“ê¸€ì˜ ìˆ˜ë¡œ (comment_count) ì˜ìƒ ë°˜ì‘ì— ëŒ€í•œ íŒë‹¨ì„ í•  ìˆ˜ ìˆë‹¤. viewcountëŒ€ë¹„ ëŒ“ê¸€ìˆ˜ê°€ ê°€ì¥ ë†’ì€ ì˜ìƒì„ í™•ì¸í•˜ë¼ (view_countê°’ì´ 0ì¸ ê²½ìš°ëŠ” ì œì™¸í•œë‹¤)
"""
```

```python
target = df.loc[df['view_count'] > 0].copy()

def getRatio(x):
    return x['comment_count'] / x['view_count']

target['ratio'] = target.apply(getRatio, axis=1)
ans = target.sort_values('ratio', ascending=False).iloc[0].title
ans
```

```python
"""
Question 7

ëŒ“ê¸€ì˜ ìˆ˜ë¡œ (comment_count) ì˜ìƒ ë°˜ì‘ì— ëŒ€í•œ íŒë‹¨ì„ í•  ìˆ˜ ìˆë‹¤.viewcountëŒ€ë¹„ ëŒ“ê¸€ìˆ˜ê°€ ê°€ì¥ ë‚®ì€ ì˜ìƒì„ í™•ì¸í•˜ë¼ (view_counts, ratioê°’ì´ 0ì¸ê²½ìš°ëŠ” ì œì™¸í•œë‹¤.)
"""
```

```python
target = df.loc[df['view_count'] > 0].copy()
target['ratio'] = target['comment_count'] / target['view_count']
target = target[target.ratio > 0]
ans = target.sort_values('ratio').iloc[0].title
ans
```
